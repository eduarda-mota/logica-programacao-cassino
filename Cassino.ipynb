```python
# Função que devolve o módulo de x

def fabs(x):
  if x < 0:
    x = -x
  return x
```

```python
# Função que devolve um inteiro, o qual não é maior que x

def floor(x):
    return int(x)
```

```python
# Função que devolve o seno de x

def seno(x):
    termo = x
    soma = x
    i = 1
    while abs(termo) >= 0.00000001:
        soma = soma + termo
        termo = -(termo*x*x)/((2*i)*(2*i+1))
        i = i + 1
    return soma
```

```python
# Código Principal

print('Bem vindo (a) à simulação do jogo das máquinas de "Seven n Half"\n')
inteiro = int(input("Digite um número inteiro para começar:\n")) # Inteiro dado pelo usuário para começar o sorteio
print("\n")
auxinteiro = inteiro # Variável auxiliar para guardar o inteiro dado pelo usuário
dig = 1 # Guarda a quantidade de dígitos do número

while auxinteiro != 0: # Descobre quantos dígitos o inteiro dado pelo usuário tem
    dig = 10*dig
    auxinteiro = auxinteiro/10

caixa = float(inteiro/dig)

teto = float(0) # Guarda o valor do teto

while teto <= 7.5: # Loop para percorrer todos os tetos possíveis
  derrotasbanc = 0 # para cada teto, o número de derrotas da banca deve ser zerado
  i = 1
  while i <= 100: # Loop para realizar as 100 simulações pedidas
    pontbanc = 0 # para cada simulação, a pontuação da banca deve ser zerada
    pontapost = 0 # para cada simulação, a pontuação do apostador deve ser zerada
    count1 = 0 # para cada simulação, o número de cartas do apostador deve ser zerado
    count2 = 0 # para cada simulação, o número de cartas da banca deve ser zerado
    while pontapost < teto: # cartas serão sorteadas para o apostador enquanto sua pontuação for menor que o teto
        rifa = float((9821.0*fabs(seno(caixa))) + 0.211327) # variável do sorteio
        caixa = rifa - floor(rifa)
        carta = floor(caixa*10 + 1) # carta sorteada (entre 1 e 10)
        count1 = count1 + 1 # contagem de cartas sorteadas para o apostador
        if carta==8 or carta==9 or carta==10:
            pontapost = pontapost + 0.5
        else:
            pontapost = pontapost + carta # de acordo com a carta sorteada, um valor é somado à pontuação do apostador
    if pontapost <= 7.5: # se a pontuação do apostador não ultrapassou 7.5, a banca irá sortear cartas para ela
        ganhei = 1
        while pontbanc <= pontapost and pontbanc < 7.5 and ganhei== 1: # cartas para a banca serão sorteadas enquanto sua pontuação for menor ou igual a pontuação do apostador, mas também menor que 7.5
            rifa = float((9821.0*fabs(seno(caixa))) + 0.211327)
            caixa = rifa - floor(rifa)
            carta = floor(caixa*10 + 1) # carta sorteada
            count2 = count2 + 1 # contagem de cartas sorteadas para a banca
            if carta==8 or carta==9 or carta==10:
                pontbanc = pontbanc + 0.5
            else:
                pontbanc = pontbanc + carta # de acordo com a carta sorteada, um valor é somado à pontuação da banca
            if pontbanc == pontapost and count2 <= count1: # a banca ganha se tiver a mesma pontuação que o apostador ao mesmo tempo em que possui menos ou o mesmo número de cartas
                ganhei = 0
        if (pontbanc == pontapost and count2 > count1) or pontbanc > 7.5:  # verificando se a banca perdeu
            derrotasbanc = derrotasbanc + 1
    i = i + 1
  print(f"{teto}  {derrotasbanc}  ") # printando as derrotas da banca correspondentes a cada teto
  j = 1
  while j <= derrotasbanc:
    print("*", end = " ")
    j = j + 1
  print(" \n")
  teto = teto + 0.5
```
